# Atividade

Nesta atividade avaliativa, desenvolveremos um pequeno projeto para analisar dados do Painel de COVID
do Ministério da Saúde.

**Item 1.** Importe os dados de COVID-19 do ano de 2021, obtidos no site do
Ministério da Saúde, ou baixada da página do curso.

```python
dados = pd.read_excel('/content/drive/MyDrive/PosDataScience/dados_covid19_2021.xlsx')
dados.head()
```

**Item 2.** Realize a limpeza dos dados de todas as colunas (NA’s, Datas, etc).
Use as funções e as ferramentas trabalhadas em aula.

```python
dados.fillna(value='', inplace=True)
dados.info()
```

**Item 3.** Use a função “factor” para categorizar a coluna “Metro/Interior” no
data frame.

```python
factor = pd.Categorical(dados["Metro/Interior"]) # função no Python semelhante ao factor no R
dados["Metro/Interior"] = factor.codes
dados.head()
```

**Item 4.** Realize um resumo dos dados utilizando a função summary.

```python
dados.describe()
```

**Item 5.** Filtre os dados para a cidade de Goiânia e mostre as 10 primeiras linhas
do data frame.

```python
# filtra com o comando .loc passando a coluna do filtro e o valor. A função sample pega as 10 primeiras linhas do data frame
amostra_goiania = dados.loc[dados.Município == 'Goiânia']
amostra_goiania.reset_index(inplace=True)
amostra_goiania.sample(10)
```

**Considerando os dados Filtrados de Goiânia**

**Item 6.** Calcule e exiba o número de casos e de óbitos acumulados de COVID-19
no ano de 2021.

```python
amostra_goiania[['semana', 'ano']] = amostra_goiania['Ano_Semana'].str.split('/', expand=True)

amostra_goiania.semana = amostra_goiania.semana.astype(int)

item6 = amostra_goiania.loc[amostra_goiania.ano == '2021']
item6[['Casos Acumulados', 'Óbitos Acumulados']].sum()
```

**Item 7.** Gere e exiba um gráfico de linha de casos acumulados de COVID-19 por
semana em 2021.

```python
sns.lineplot(data=item6, x='semana', y='Casos Acumulados')
```

**Item 8.** Gere e exiba um gráfico de linha de óbitos acumulados de COVID-19
por semana em 2021.

```python
sns.lineplot(data=item6, x='semana', y='Óbitos Acumulados')
```

**Item 9.** Crie uma nova coluna denominada “Regiao”, partir da coluna “UF”.
Agrupe as UF segundo suas regiões específicas: Norte, Nordeste, Centro-Oeste,
Sul e Sudeste.

```python
regioes_brasil = {
    "Norte": ["AC", "AP", "AM", "PA", "RO", "RR", "TO"],
    "Nordeste": ["AL", "BA", "CE", "MA", "PB", "PE", "PI", "RN", "SE"],
    "Centro-Oeste": ["DF", "GO", "MT", "MS"],
    "Sudeste": ["ES", "MG", "RJ", "SP"],
    "Sul": ["PR", "RS", "SC"]
}

dados.loc[:, 'Regiao'] = dados.UF.apply(lambda x: [key for key, value in regioes_brasil.items() if x in value][0])
dados.head()
```

```python
dados.Regiao.value_counts().reset_index()
```
